```python

import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec

np.random.seed(42)

plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Times New Roman']
plt.rcParams['axes.labelweight'] = 'normal'
```

```python


# --- 1. 生成模拟数据 ---
# 您可以随意更改这里的均值和标准差，图表会自动适应
np.random.seed(42)
lb_carbide_avg, lb_carbide_std = 0.14, 0.02
tm_carbide_avg, tm_carbide_std = 0.14, 0.02
lb_carbide_data = np.random.normal(lb_carbide_avg, lb_carbide_std, 450)
tm_carbide_data = np.random.normal(tm_carbide_avg, tm_carbide_std, 150)

lb_k_avg, lb_k_std = 0.38, 0.07
tm_k_avg, tm_k_std = 0.34, 0.06
lb_k_data = np.random.normal(lb_k_avg, lb_k_std, 450)
tm_k_data = np.random.normal(tm_k_avg, tm_k_std, 150)

# --- 2. 动态计算所有子图的坐标轴范围 ---

# 为 Carbide Percentage (a, c, e) 计算范围
combined_carbide_data = np.concatenate((lb_carbide_data, tm_carbide_data))
carbide_min, carbide_max = combined_carbide_data.min(), combined_carbide_data.max()
carbide_range = carbide_max - carbide_min
# 为散点图 (a) 计算 Y 轴范围，顶部留出更多空间给文本
carbide_scatter_ylim = (carbide_min - 0.1 * carbide_range, carbide_max + 0.3 * carbide_range)
# 为直方图 (c, e) 计算 X 轴范围
carbide_hist_xlim = (carbide_min - 0.05 * carbide_range, carbide_max + 0.05 * carbide_range)
# 为直方图 (c, e) 计算 Y 轴范围
lb_carbide_counts, _ = np.histogram(lb_carbide_data, bins=20)
tm_carbide_counts, _ = np.histogram(tm_carbide_data, bins=20)
carbide_max_count = max(lb_carbide_counts.max(), tm_carbide_counts.max())
carbide_hist_ylim = (0, carbide_max_count * 1.1)

# 为 k (b, d, f) 计算范围
combined_k_data = np.concatenate((lb_k_data, tm_k_data))
k_min, k_max = combined_k_data.min(), combined_k_data.max()
k_range = k_max - k_min
# 为散点图 (b) 计算 Y 轴范围
k_scatter_ylim = (k_min - 0.1 * k_range, k_max + 0.3 * k_range)
# 为直方图 (d, f) 计算 X 轴范围
k_hist_xlim = (k_min - 0.05 * k_range, k_max + 0.05 * k_range)
# 为直方图 (d, f) 计算 Y 轴范围
lb_k_counts, _ = np.histogram(lb_k_data, bins=25)
tm_k_counts, _ = np.histogram(tm_k_data, bins=25)
k_max_count = max(lb_k_counts.max(), tm_k_counts.max())
k_hist_ylim = (0, k_max_count * 1.1)

# --- 3. 创建图形和子图布局 ---
fig = plt.figure(figsize=(12, 12))
gs = gridspec.GridSpec(3, 2, figure=fig, hspace=0.4, wspace=0.3)
ax1 = fig.add_subplot(gs[0, 0])
ax2 = fig.add_subplot(gs[0, 1])
ax3 = fig.add_subplot(gs[1, 0])
ax4 = fig.add_subplot(gs[1, 1])
ax5 = fig.add_subplot(gs[2, 0])
ax6 = fig.add_subplot(gs[2, 1])

# --- 4. 绘制子图 (a): Carbide Percentage 散点图 ---
ax1.scatter(range(len(lb_carbide_data)), lb_carbide_data, facecolors='none', edgecolors='royalblue', label='LB', marker='o')
ax1.scatter(range(len(tm_carbide_data)), tm_carbide_data, facecolors='none', edgecolors='orchid', label='TM', marker='^')
ax1.axhline(y=np.mean(lb_carbide_data), color='royalblue', linestyle='-')
ax1.axhline(y=np.mean(tm_carbide_data), color='orchid', linestyle='--')
ax1.set_title('(a)', loc='left', fontsize=16)
ax1.set_xlabel('Image Index', fontsize=14)
ax1.set_ylabel('Carbide Percentage', fontsize=14)
ax1.text(0.05, 0.95, f'LB:\nAverage:{np.mean(lb_carbide_data):.2f}  STD:{np.std(lb_carbide_data):.2f} \nTM:\nAverage:{np.mean(tm_carbide_data):.2f}  STD:{np.std(tm_carbide_data):.2f}', transform=ax1.transAxes, fontsize=12, verticalalignment='top')

ax1.tick_params(axis='both', which='major', labelsize=12)

# --- 5. 绘制子图 (b): k 值散点图 ---
ax2.scatter(range(len(lb_k_data)), lb_k_data, facecolors='none', edgecolors='royalblue', label='LB', marker='o')
ax2.scatter(range(len(tm_k_data)), tm_k_data, facecolors='none', edgecolors='orchid', label='TM', marker='^')
ax2.axhline(y=np.mean(lb_k_data), color='royalblue', linestyle='-')
ax2.axhline(y=np.mean(tm_k_data), color='orchid', linestyle='--')
ax2.set_title('(b)', loc='left', fontsize=16)
ax2.set_xlabel('Image Index', fontsize=14)
ax2.set_ylabel('k', fontsize=14)
ax2.text(0.05, 0.95, f'LB:\nAverage:{np.mean(lb_k_data):.2f}  STD:{np.std(lb_k_data):.2f} \nTM:\nAverage:{np.mean(tm_k_data):.2f}  STD:{np.std(tm_k_data):.2f}', transform=ax2.transAxes, fontsize=12, verticalalignment='top')

ax2.tick_params(axis='both', which='major', labelsize=12)

# --- 6. 绘制直方图 (c, e, d, f) ---
# ... (这部分绘图代码与之前完全相同，此处省略以保持简洁)
ax3.hist(lb_carbide_data, bins=20, color='royalblue', alpha=0.7, label='LB', edgecolor='black', linewidth=1.2)
ax3.set_title('(c)', loc='left', fontsize=16)
ax3.set_ylabel('Image number', fontsize=14)
ax3.legend(fontsize=12)
ax3.tick_params(axis='both', which='major', labelsize=12)
ax5.hist(tm_carbide_data, bins=20, color='orchid', alpha=0.7, label='TM', edgecolor='black', linewidth=1.2)
ax5.set_title('(e)', loc='left', fontsize=16)
ax5.set_xlabel('Carbide percentage', fontsize=14)
ax5.set_ylabel('Image number', fontsize=14)
ax5.legend(fontsize=12)
ax5.tick_params(axis='both', which='major', labelsize=12)
ax4.hist(lb_k_data, bins=25, color='royalblue', alpha=0.7, label='LB', edgecolor='black', linewidth=1.2)
ax4.set_title('(d)', loc='left', fontsize=16)
ax4.set_ylabel('Image number', fontsize=14)
ax4.legend(fontsize=12)
ax4.tick_params(axis='both', which='major', labelsize=12)
ax6.hist(tm_k_data, bins=25, color='orchid', alpha=0.7, label='TM', edgecolor='black', linewidth=1.2)
ax6.set_title('(f)', loc='left', fontsize=16)
ax6.set_xlabel('k', fontsize=14)
ax6.set_ylabel('Image number', fontsize=14)
ax6.legend(fontsize=12)
ax6.tick_params(axis='both', which='major', labelsize=12)

# --- 7. 应用统一且动态的坐标轴范围 ---
# 应用范围到子图 (a)
ax1.set_ylim(carbide_scatter_ylim)

# 应用范围到子图 (b)
ax2.set_ylim(k_scatter_ylim)

# 应用范围到子图 (c) 和 (e)
ax3.set_xlim(carbide_hist_xlim)
ax5.set_xlim(carbide_hist_xlim)
ax3.set_ylim(carbide_hist_ylim)
ax5.set_ylim(carbide_hist_ylim)

# 应用范围到子图 (d) 和 (f)
ax4.set_xlim(k_hist_xlim)
ax6.set_xlim(k_hist_xlim)
ax4.set_ylim(k_hist_ylim)
ax6.set_ylim(k_hist_ylim)

# --- 8. 显示图形 ---
plt.show()
```
